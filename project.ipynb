## *Observation:*

* "M"
* "k"
* "Varies with devices"   





```python
def clean_size_column(size):         # Function to clean and convert 'Size' column
  if 'M' in size: # converting megabytes(M)
    return float(size.replace('M',''))
  elif 'K' in size:   # converting kilobytes(K) to megabytes(M)
    return float(size.replace('K',''))/1024
  else:          # handles varies with devices or unexpected values
    return np.nan
```

```python
df['Size']=df['Size'].apply(clean_size_column) # apply the function on 'Size' column
df.dropna(subset=['Size'])        # drop the value contains nan
```

***`Which genres have the most apps with over 1 million installs?`***

```python
def genres_topmost_app(df,min_installs=1000000,top=5):
  filtered_df=df[df['Installs'] > min_installs]
  # Assign filtered_df to grouped_df before using it
  grouped_df=filtered_df.groupby(['Genres'])[['App','Installs']].agg({'App':'count','Installs':'sum'})
  grouped_df=grouped_df.sort_values(by=['App','Installs'],ascending=False)
  return grouped_df.head(top)

topmost_app=genres_topmost_app(df,min_installs=1000000,top=5)
print(topmost_app)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(topmost_app.index, topmost_app['App'], color='skyblue')
plt.xlabel('Genres',fontsize=14,fontstyle='italic',color='brown')
plt.ylabel('Number of Apps with Over 1 Million Installs',fontsize=14,fontstyle='italic',color='brown')
plt.title('Top Genres with Most Apps Having Over 1 Million Installs',fontweight='bold',fontsize=14,fontstyle='italic',color='brown')
plt.xticks(rotation=45, ha='right')
plt.grid(linestyle='--',color='grey')
plt.tight_layout()
plt.show()
```

*`The visualization highlights that the 'Action' genre boasts the highest number of apps (nearly 160) with over 1 million installs, followed by 'Tools' with approximately 95 apps. ***`How frequently do apps get updated? Calculate the average time between updates.`***

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'])
```

```python
def update_frequency(df):
  grouped_df=df.groupby('App')['Last Updated'].apply(lambda x:x.sort_values()).reset_index()
  grouped_df['Difference']=grouped_df.groupby('App')['Last Updated'].diff()
  grouped_df=grouped_df[grouped_df['Difference'].notna()]
  return grouped_df['Difference'].mean()

average_time_updates= update_frequency(df)
print(f"The Average time between app updates are {average_time_updates}")
```

***` Analyze the trend of app updates over time. Are there any noticeable patterns or seasonal trends?`***


```python
def app_updates_overtime(df):
  df['Month']=df['Last Updated'].dt.month
  df['Year']=df['Last Updated'].dt.year
  groupby_df=df.groupby(['Year','Month']).agg({'App':'count'})
  groupby_df=groupby_df.sort_values(by='App',ascending=False).reset_index()
  return groupby_df

app_updates=app_updates_overtime(df)
print(app_updates)
```

```python
plt.figure(figsize=(9, 6))
ax=plt.gca()
ax.set_facecolor('lightgrey')
# Create a new column 'Year-Month' by formatting 'Year' and 'Month' columns of the DataFrame
app_updates['Year-Month'] = app_updates['Year'].astype(str) + '-' + app_updates['Month'].astype(str).str.zfill(2)
# Use the newly created column for plotting
plt.plot(app_updates['Year-Month'], app_updates['App'],color='blue',marker='.',linewidth=2)
plt.xlabel('Year-Month',fontsize=18,fontstyle='italic',color='Purple',fontweight='bold')
plt.ylabel('Number of Updates',fontsize=18,fontstyle='italic',color='Purple',fontweight='bold')
plt.title('App Update Trend Over Time',fontsize=19,fontstyle='italic',color='Purple',fontweight='bold')
plt.grid(True,color='grey',linestyle='dashed',alpha=0.8)
plt.xticks(rotation=102,ha='right')
plt.tight_layout()
plt.show()
```

*`This trend reflects a dynamic market where apps are actively updated at their peak popularity or launch phase, followed by reduced updates as they stabilize. Developers might prioritize newer apps or high-performing apps for continuous improvement, while older or less popular apps see fewer updates over time.`*









***` How does the average rating of apps change with the number of installs? Create a binned analysis.`***

```python
bins = [0, 1000, 10000, 100000, 1000000, 10000000, 100000000, np.inf]    # Define bins for the Installs column
labels = ['0-1k', '1k-10k', '10k-100k', '100k-1M', '1M-10M', '10M-100M', '100M+']
df['Install_Bins'] = pd.cut(df['Installs'], bins=bins, labels=labels, include_lowest=True) # Create a new column for the bins
binned_analysis = df.groupby('Install_Bins',observed=False)['Rating'].mean().reset_index()  # Calculate the average rating for each bin
binned_analysis.columns= ['Install Range', 'Average Rating']   # Rename columns for better clarity
print(binned_analysis)
```

```python
plt.figure(figsize=(9,6))
ax=plt.gca()
ax.set_facecolor('grey')
plt.barh(binned_analysis['Install Range'],binned_analysis['Average Rating'],color='orange',edgecolor='black',linewidth=1)
plt.xlabel('Average Rating',fontsize=14,fontstyle='italic')
plt.ylabel('Install Range',fontsize=14,fontstyle='italic')
plt.title('Average Rating by Install Range',fontsize=18,fontstyle='italic',fontweight='bold')
plt.grid(axis='x',linestyle='--',alpha=0.5,color='white',linewidth=1)
for i, v in enumerate(binned_analysis['Average Rating'].round(3)):
    plt.text(v + 0.5,i,str(v), color='black',fontsize=10, va='center')  # Add value labels to the bars
plt.tight_layout()
plt.show()
```

`From the analysis, we observe a positive relationship between the number of installs and the average app rating. Apps with fewer installs (e.g., 1k–10k) have slightly lower average ratings (4.04), while those with higher installs (e.g., 10M–100M) exhibit significantly higher average ratings (4.38). This trend indicates that apps with more installs tend to receive higher user satisfaction, potentially due to better quality, updates, and user trust.`
